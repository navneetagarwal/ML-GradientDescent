```python
import pandas as pd
```

```python
df = pd.read_csv('test_scores.csv')
print(df)
```

```python
x_values = df['math'].values
y_values = df['cs'].values
```

```python
import math
def gradient_descent(x, y):
    m = 0.0
    b = 0.0
    iterations = 1000000
    learning_rate = 0.0002
    n = len(x)
    prev_cost = 0.0
    for i in range(0, iterations):
        y_new = m * x + b
        m_differential = 1 / n * sum(-2 * x * (y - y_new))
        b_differential = 1 / n * sum(-2 * (y - y_new))
        error = sum([z*z for z in (y - (m * x + b)) * (y - (m * x + b))]) / n / n
        new_cost = 1 / n * sum((y - y_new) * (y - y_new))
        if math.isclose(prev_cost, new_cost, rel_tol=1e-4):
            break
        m = m - learning_rate * m_differential
        b = b - learning_rate * b_differential
        prev_cost = new_cost
        
        print(f"Slope = {m}, Intercept = {b}, Cost = {error}, mdiff = {m_differential}, b_diff = {b_differential}")
```

```python
gradient_descent(x_values, y_values)
```
